```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
sent = [
    'The quick brown fox jumps over the lazy dog.',
    'Natural language processing is a fascinating field of AI.',
    'Deep learning models require a lot of data to perform well.',
    'Word embeddings capture semantic meaning of words.',
    'Keras makes building neural networks simple and intuitive.'
]
sent
```

```python
## define vocab ulary size
vocab_size = 10000
```

```python
## representation (One Hot)
one_hot_repr = [one_hot(words, vocab_size)for words in sent]
one_hot_repr
```

```python
## word embedding reptresentation
from tensorflow.keras.layers import Embedding
# from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.utils import pad_sequences
from tensorflow.keras.models import Sequential

import numpy
```

```python
sent_len = 15

embedding_docs = pad_sequences(one_hot_repr, maxlen=sent_len, padding='pre')
embedding_docs
```

```python
## feature representation
dim = 12
```

```python
model = Sequential()
model.add(Embedding(vocab_size, dim, input_length=sent_len))
model.compile('adam', 'mse')
model.summary()
```

```python
model.predict(embedding_docs)
```
