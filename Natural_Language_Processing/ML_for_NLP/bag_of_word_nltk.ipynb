```python
import pandas as pd
```

```python
messages = pd.read_csv('spam.csv')
messages.head()
```

```python
messages.info()
```

```python
messages = messages.dropna(axis=1, how='any')
```

```python
messages.info()
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import re

porterstemmer = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(messages)):
    message = re.sub('[^a-zA-Z]', ' ', messages['v2'][i])
    message = message.lower()
    message = message.split()
    message = [porterstemmer.stem(word) for word in message if not word in set(stopwords.words('english'))] # you can use lemmatizer or snowballstemmer
    message = ' '.join(message)
    corpus.append(message)
```

```python
corpus[:5]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=2500, binary=True) # binary true for binary bag of words
```

```python
cv.fit_transform(corpus).toarray()
```
